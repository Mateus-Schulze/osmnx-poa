```python
import networkx as nx
import osmnx as ox

%matplotlib inline
ox.__version__
```

```python
# Definir ponto de localização
# location_point_1 (rua Duque de Caxias, Porto Alegre)

location_point_1 = (-30.0333469, -51.2319421)
```

```python
# Criar rede apenas de nós dentro de 500m caminhando ao longo da rede a partir do ponto
G = ox.graph_from_point(location_point_1, dist=500, dist_type="network", network_type="walk")
fig = ox.plot_graph(G, node_color="red", bgcolor="white", node_size=10, edge_linewidth=1, edge_color="black")
```
